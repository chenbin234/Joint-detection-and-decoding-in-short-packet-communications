```python
# plot y = log2(1+x) for x in [-5, 20]

import numpy as np
import matplotlib.pyplot as plt

x_db = np.linspace(-5, 20, 100)
x_lin = 10 ** (x_db / 10)
y = np.log2(1 + x_lin)

plt.plot(x_db, y)
plt.xlabel('x')
# set x-axis from -5 to 20
plt.xlim(-5, 20)
plt.ylabel('y')
plt.ylim(0, 7)
plt.title('y = log2(1+x)')
plt.grid(True)
plt.show()
```
